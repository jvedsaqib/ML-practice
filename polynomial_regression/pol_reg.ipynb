```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from scipy import stats
```

```python
df = pd.read_csv("asset/salary.csv")
df.head(5)
```

```python

slope, intercept, r, p, std_err = stats.linregress(df.Level, df.Salary)

def mapFun(x):
    return slope * x + intercept

mymodel = list(map(mapFun, df.Level))

%matplotlib inline
plt.plot(df.Level, mymodel)
plt.scatter(df.Level, df.Salary)
```

```python

mymodel = np.poly1d(np.polyfit(df.Level, df.Salary, 3))
myline = np.linspace(1, 10, 100)

%matplotlib inline
plt.plot(myline, mymodel(myline))
plt.scatter(df.Level, df.Salary)
```
